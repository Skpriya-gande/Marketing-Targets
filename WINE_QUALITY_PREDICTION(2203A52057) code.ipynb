```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

class LogisticRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, Y):
        m, n = X.shape
        self.weights = np.zeros(n)
        self.bias = 0

        for _ in range(self.num_iterations):
            h = self.sigmoid(np.dot(X, self.weights) + self.bias)
            dw = (1 / m) * np.dot(X.T, (h - Y))
            db = (1 / m) * np.sum(h - Y)
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        h = self.sigmoid(np.dot(X, self.weights) + self.bias)
        y_pred = np.where(h > 0.5, 1, 0)
        return y_pred

# Load your data, and perform necessary preprocessing
# Assuming 'p' contains your data
p['goodquality'] = [1 if x >= 7 else 0 for x in p['quality']]
X = p.drop(['quality', 'goodquality'], axis=1)
Y = p['goodquality']

# Perform PCA for dimensionality reduction
n_components = 2
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)

# Split the PCA-transformed data into training and testing sets (70% training, 30% testing)
split_ratio = 0.7
X_train, X_test, Y_train, Y_test = train_test_split(X_pca, Y, test_size=1 - split_ratio, random_state=7)

# Create an instance of the LogisticRegression class
model = LogisticRegression(learning_rate=0.01, num_iterations=1000)

# Fit the model on the training data
model.fit(X_train, Y_train)

# Make predictions on the test data
predictions = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, predictions)
report = classification_report(Y_test, predictions)
conf_matrix = confusion_matrix(Y_test, predictions)

print(f"Accuracy: {accuracy}")
print(report)

# Display the confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for logisticRegression')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import classification_report

class Perceptron:
    def __init__(self, learning_rate=0.1, n_iters=10000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.activation_func = self._unit_step_func
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize parameters
        self.weights = np.zeros(n_features)
        self.bias = 0

        y_ = np.array([1 if i > 5 else 0 for i in y])

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)

                # Perceptron update rule
                update = self.lr * (y_[idx] - y_predicted)

                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        return y_predicted

    def _unit_step_func(self, x):
        return np.where(x >= 0, 1, 0)

def accuracy(y_true, y_pred):
    accuracy = np.sum(y_true == y_pred) / len(y_true)
    return accuracy

# Create a synthetic dataset
x, y = datasets.make_blobs(
    n_samples=150, n_features=2, centers=2, cluster_std=1.05, random_state=2
)
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=123
)

# Instantiate and train the perceptron with adjusted parameters
p = Perceptron(learning_rate=0.1, n_iters=10000)
p.fit(X_train, y_train)

# Make predictions
predictions = p.predict(X_test)

# Calculate accuracy
accuracy = accuracy(y_test, predictions)
print("Perceptron classification accuracy:", accuracy)

# Generate a classification report
report = classification_report(y_test, predictions)
print("Classification Report:\n", report)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.utils import resample
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_csv(r"/content/WineQT.csv")

# Classify the wine as 'goodquality' if its quality rating is 7 or higher
data['goodquality'] = [1 if x >= 7 else 0 for x in data['quality']]
x = data.drop(['quality', 'goodquality'], axis=1)
Y = data['goodquality']

# Split the data into training and testing sets (70% training, 30% testing)
split_ratio = 0.7
X_train, X_test, Y_train, Y_test = train_test_split(x, Y, test_size=1 - split_ratio, random_state=7)

# Define the number of bootstrap iterations
n_iterations = 100  # Adjust this number based on your computational resources

# Initialize a list to store accuracy scores from each iteration
accuracy_scores = []

# Perform bootstrap sampling
for i in range(n_iterations):
    # Resample the training data with replacement
    X_resampled, Y_resampled = resample(X_train, Y_train)

    # Create a new instance of the Random Forest classifier
    model = RandomForestClassifier(n_estimators=100, random_state=42)

    # Fit the model on the resampled training data
    model.fit(X_resampled, Y_resampled)

    # Make predictions on the test data
    pred_y = model.predict(X_test)

    # Calculate accuracy and store it in the list
    accuracy = accuracy_score(Y_test, pred_y)
    accuracy_scores.append(accuracy)

# Calculate the mean and standard deviation of accuracy scores
mean_accuracy = np.mean(accuracy_scores)
std_accuracy = np.std(accuracy_scores)

print(f"Mean Accuracy: {mean_accuracy:.4f}")
print(f"Standard Deviation of Accuracy: {std_accuracy:.4f}")

# Plotting the distribution of accuracy scores
plt.figure(figsize=(10, 6))
plt.hist(accuracy_scores, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Accuracy Scores (Bootstrap)')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with random data for demonstration purposes
data = pd.DataFrame({
    'value': np.random.normal(loc=0, scale=1, size=1000)
})

# Bootstrap function
def bootstrap(data, n_bootstrap):
    bootstrap_samples = np.zeros(n_bootstrap)
    n = len(data)
    for i in range(n_bootstrap):
        sample = np.random.choice(data, size=n, replace=True)
        bootstrap_samples[i] = np.mean(sample)
    return bootstrap_samples

# Perform 1000 bootstrap samples
bootstrap_means = bootstrap(data['value'], 1000)

# Estimate the mean and calculate the 95% confidence interval
estimated_mean = np.mean(bootstrap_means)
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])

print(f"Estimated Mean: {estimated_mean}")
print(f"95% Confidence Interval: {confidence_interval}")

# Plotting
plt.hist(bootstrap_means, bins=30, color='skyblue', edgecolor='black')
plt.axvline(x=confidence_interval[0], color='red', linestyle='--', label='95% CI')
plt.axvline(x=confidence_interval[1], color='red', linestyle='--')
plt.axvline(x=estimated_mean, color='green', linestyle='-', label='Estimated Mean')
plt.title('Bootstrap Distribution of Sample Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
